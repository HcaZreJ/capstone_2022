# Local data retrieval


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
# Data importing
df = pd.read_csv('data_v2.csv', header=0, index_col=0, dtype=str,
                 parse_dates=['publishedAt'], infer_datetime_format=True)

embeddings = np.load("embeddings_v2.npy")
similarity_matrix = np.load("similarity_matrix_v2.npy")
```

```python
# Plotting every vector as a histogram of its dimension values
for i in range(embeddings.shape[0]):
    plt.hist(embeddings[i,:], density=True, bins=20, alpha=0.5)

plt.xlabel("Dimension value")
plt.ylabel("Probability Density")
plt.title("Every news vector plotted as a histogram")
plt.show()
```

```python
# Checking each dimension's variance
for i in range(embeddings.shape[1]):
    plt.hist(embeddings[:,i], density=True, bins=20, alpha=0.5)

plt.xlabel("Dimension value")
plt.ylabel("Probability Density")
plt.title("Every dimension's histogram")
plt.show()
```
